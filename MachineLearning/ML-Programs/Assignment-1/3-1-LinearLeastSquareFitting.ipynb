```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Salary_Data.csv')
X = data.iloc[:,0].values
y = data.iloc[:,-1].values
```

```python
def linearfitting(x,y):
    m = b = 0       # initializing slope and intercept to zero
    n = len(X)
    learning_rate = 0.001
    accuracy = 1
    condition = True
    i = 1
    while(accuracy>10**(-9)):

        y_predicted = m*x + b
        dm = (-2/n)*sum(x*(y-y_predicted))
        db = (-2/n)*sum(y-y_predicted)

        m_new = m - learning_rate*dm
        b_new = b - learning_rate*db
        
        accuracy = max(abs(m_new-m),abs(b_new-b))
        
        print(f'Iteration = {i}     m = {m:.15f}    b = {b:.15f}    accuracy = {accuracy:.15f}')
        i +=1
        m = m_new
        b = b_new

    return m,b
```

```python
m, b = linearfitting(X,y)
```

```python
ypred = m*X + b
```

```python
plt.scatter(X,y, marker = '.')
plt.plot(X,ypred,color = 'red')
plt.title('Linear Least Square Fitting')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show
```

```python
error = abs(y[0]-ypred[0])/y[0]
for i in range(1,len(X)):
    e = abs(y[i]-ypred[i])/y[i]
    if(error<e):
        error=e
print(error)
```
