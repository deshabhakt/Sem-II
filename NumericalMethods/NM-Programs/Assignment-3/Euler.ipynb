```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def euler(x0,xn,y0,n,f,g):
    lst = [[x0,y0,g(x0)]]
    h = (xn-x0)/n
    print('x0\t\ty_approx\ty_exact\n')
    for i in range(n):
        yn = y0 + h * f(x0, y0)
        
        print(f'{x0:f}\t{y0:f}\t{g(x0):f}') 
        y0 = yn
        x0 = x0+h
        lst.append([x0,y0,g(x0)])
    print(f'{x0:f}\t{y0:f}\t{g(x0):f}') 
    return lst
```

```python
def eulerModified(x0,xn,y0,n,f,g):
    lst = [[x0,y0,g(x0)]]
    h = (xn-x0)/n
    print('x0\t\ty_approx\ty_exact\n')
    for i in range(n):
        k = h*f(x0, y0)
        yn = y0 + h * f(x0 +h/2, y0+k/2)
        
        print(f'{x0:f}\t{y0:f}\t{g(x0):f}') 
        y0 = yn
        x0 = x0+h
        lst.append([x0,y0,g(x0)])
    print(f'{x0:f}\t{y0:f}\t{g(x0):f}') 
    return lst
```

```python
def eulerImproved(x0,xn,y0,n,f,g):
    lst = [[x0,y0,g(x0)]]
    h = (xn-x0)/n
    print('x0\t\ty_approx\ty_exact\n')
    for i in range(n):
        k = h*f(x0, y0)
        yn = y0 + (h/2) *( f(x0, y0) + f(x0 + h, y0 + k) )
        
        print(f'{x0:f}\t{y0:f}\t{g(x0):f}') 
        y0 = yn
        x0 = x0+h
        lst.append([x0,y0,g(x0)])
    print(f'{x0:f}\t{y0:f}\t{g(x0):f}') 
    return lst
```

```python
def rk4(x0,xn,y0,n,f,g):
    lst = [[x0,y0,g(x0)]]
    h = (xn-x0)/n
    print('x0\t\ty_approx\ty_exact\n')
    for i in range(n):
        k1 = h*f(x0, y0)
        k2 = h*f(x0+h/2,y0+k1/2)
        k3 = h*f(x0+h/2,y0+k2/2)
        k4 = h*f(x0+h,y0+k3)
        yn = y0 + (1/6) *(k1 + 2*k2 + 2*k3 + k4)
        
        print(f'{x0:f}\t{y0:f}\t{g(x0):f}') 
        y0 = yn
        x0 = x0+h
        lst.append([x0,y0,g(x0)])
    print(f'{x0:f}\t{y0:f}\t{g(x0):f}') 
    return lst
```

```python
f = lambda x,y: x*y
g = lambda x: math.exp((x**2)/2)
```

```python
a = 0
b = 2
y0 = 1

N = 20
```

```python
lst_euler = euler(a,b,y0,N,f,g)
```

```python
lst_euler_modified = eulerModified(a,b,y0,N,f,g)
```

```python
lst_euler_improved = eulerImproved(a,b,y0,N,f,g)
```

```python
x = np.linspace(a,b,1000)
y = []
for i in range(len(x)):
    y.append(g(x[i]))

x_approx = []
y_approx = []
y_exact_points = []
for i in lst_euler:
    x_approx.append(i[0])
    y_approx.append(i[1])
    y_exact_points.append(i[2])

plt.plot(x_approx,y_approx,label='Approxiamte',color='r')
plt.scatter(x_approx,y_exact_points,color='blue')
plt.plot(x,y,label='Exact',color='green')
plt.legend()
plt.show()
```
